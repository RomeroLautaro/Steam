TRABAJAMOS CON EL ARCHIVO ITEMS

1-Importamos las librerias a utilizar

```python
import pandas as pd
import json
import ast
```

2-Abrimos el archivo en formato json

```python
# Ruta al dataset australian_user_reviews
items = 'D:/escritorio/STEAM/Json/australian_users_items.json'

# Se lee de cada línea del dataset
filas_items = []
with open(items) as f:
    for line in f.readlines():
        filas_items.append(ast.literal_eval(line))

# Se convierte en dataframe
df_items = pd.DataFrame(filas_items)
df_items
```

REALIZAMOS TRATAMIENTOS A LA COLUMNAS ITEMS

```python
# se normaliza la columna 'items'
df_items2 = pd.json_normalize(filas_items, record_path=['items'], meta=['steam_id','items_count','user_id', 'user_url'] )
df_items2
```

Verificar los valores nulos

```python
# Obtener la cantidad de valores nulos por columna
valores_nulos_por_columna = df_items2.isnull().sum()

# Calcular el porcentaje de valores nulos por columna
total_valores_por_columna = len(df_items2)
porcentaje_valores_nulos_por_columna = (valores_nulos_por_columna / total_valores_por_columna) * 100

# Crear un DataFrame con el formato deseado
resultado_df = pd.DataFrame({
    'nombre_campo': valores_nulos_por_columna.index,
    'no_nulos_%': 100 - porcentaje_valores_nulos_por_columna,
    'nulos_%': porcentaje_valores_nulos_por_columna,
    'nulos': valores_nulos_por_columna
})

# Mostrar el resultado
print(resultado_df)
```

Verificamos si hay valores duplicados

```python
# Devuelve los duplicados
duplicados = df_items2.loc[df_items2.duplicated()]
duplicados
```

Eliminamos los valores duplicados

```python
df_items2 = duplicados.drop_duplicates(keep='first')
df_items2
```

Eliminamos las siguientes columnas

```python
df_items2.drop(['steam_id','playtime_2weeks','user_url'],axis=1, inplace=True)
df_items2.head()
```

GUARDAMOS EL ARCHIVO EN FORMATO PARQUET

```python
df_items2.to_parquet('D:/escritorio/STEAM/parquet/item.parquet')
```
