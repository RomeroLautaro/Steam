TRABAJO AL ARCHIVO OUTPUT

Importamos las librerias a utilizar

```python
import pandas as pd
import json
```

Abrimos el archivo que esta en Formato JSON

```python
# Ruta al dataset australian_user_reviews
output_ruta = 'D:\escritorio\STEAM\Json\output_steam_games.json'

# Se lee de cada lÃ­nea del dataset
filas_games = []
with open(output_ruta) as f:
    for line in f.readlines():
        data = json.loads(line)
        filas_games.append(data)

# Se convierte en dataframe
output = pd.DataFrame(filas_games)
output
```

Consultamos cuantos valores nulos posee nuestro archivo y lo eliminamos

```python
# Cuenta la cantidad de nulos por columna
print(output.isnull().sum())
print("Total de valores nulos:",output.isnull().sum().sum())
```

```python
output.dropna(how='all', inplace=True)
```

Controlamos cuantos valores nulos hay por columna

```python
print(output.isnull().sum())
```

Consultamos cuanto valores duplicados con de la columna id

```python
output['id'].duplicated().sum()
```

```python
# Cuento las veces que aparece cada id
counts_id = output['id'].value_counts()
# Imprime los que aparecen mas de 1 vez
print(counts_id[counts_id>1])
```

```python
# Identificar valores duplicados en la columna 'ID'
duplicados = output[output.duplicated(subset=['id'], keep=False)]
# Obtener las filas completas correspondientes a los valores duplicados
filas_duplicadas = output[output['id'].isin(duplicados['id'])]
filas_duplicadas
```

```python
# Eliminar los duplicados manteniendo solo una instancia de cada uno
output_sin_duplicados = output.drop_duplicates(subset=['id'], keep='first')
output_sin_duplicados
```

Normalizamos los datos de la columna id

```python
output['id'] = output['id'].astype(str)

# Remplazo los '.0' por '' (vacio)
output['id'] = output['id'].str.replace('.0','')
# Elimino lo vacio
output['id'] = output['id'].str.strip()
```

Cambiamos los valores de la fecha a datatime

```python
output['release_date']= pd.to_datetime(output['release_date'], errors='coerce')
```

```python
# Extraigo el aÃ±o lo almaceno como string en 'release_year' y relleno con 'sin dato' los nulos
output['release_year'] = output['release_date'].dt.year.fillna('sin dato').astype(str)

# Esto va porque al pasar el aÃ±o a string aparece con .0 al final (ej: 2015.0)
output['release_year'] = output['release_year'].str.replace('.0','')
#Elimina los espacios en blanco 
output['release_year'] = output['release_year'].str.strip()
```

```python
#Elimino la columna 'release_date'
output.drop(['release_date'],axis=1,inplace=True)
```

Cambiamos los datos de la columna price

```python
precios = output['price'].unique()
precios
```

```python
# Define una funciÃ³n para convertir valores a punto flotante
def reemplaza_a_flotante(valor):
    try:
        # Intenta convertir el valor a punto flotante
        return float(valor)
    except (ValueError, TypeError):
        # Si no se puede convertir, devuelve NaN
        return float('nan')

# Aplica la funciÃ³n a la columna 'price'
output['price'] = output['price'].apply(reemplaza_a_flotante)

# Verifica el tipo de datos de la columna 'price'
print(output['price'].dtype)
```

```python
output['price'].fillna(0.00, inplace=True)
```

Transformacion de la columna genre

```python
# Funcion para obtener  
def clean_genres(genres):
    if isinstance(genres, list):# Si 'genres' es una lista. Devuelve el primer elemento
        return genres[0]
    elif isinstance(genres, str): # Si 'genres' es un string. Devuelve ese valor
        return genres
    else:
        return 'Pending classification' 

# Se aplica la funcion a la columna 'genres'  
output['genres'] = output['genres'].apply(clean_genres)
```

```python
output['genres'].value_counts()
```

Volvemos a ver cuanto nulos hay por columnas

```python
# Cuento la cantidad de nulos que quedaron por columna
print(output.isnull().sum())
```

Eliminamos las columnas que no vamos a necesitar para llevar adelante el proyecto

```python
output_games = output.drop(['tags', 'specs', 'url', 'reviews_url', 'publisher', 'title', 'early_access'], axis=1)
output_games.columns
```

Eliminamos los valores nulos de la columna developer

```python
# Eliminar filas con valores nulos en la columna 'developer'
output_games = output_games.dropna(subset=['developer'])
```

Se verifican el tipo de dato y si quedan nulos.

```python
# Obtener la cantidad de valores nulos por columna
valores_nulos_por_columna = output_games.isnull().sum()

# Calcular el porcentaje de valores nulos por columna
total_valores_por_columna = len(output_games)
porcentaje_valores_nulos_por_columna = (valores_nulos_por_columna / total_valores_por_columna) * 100

# Crear un DataFrame con el formato deseado
resultado_df = pd.DataFrame({
    'nombre_campo': valores_nulos_por_columna.index,
    'no_nulos_%': 100 - porcentaje_valores_nulos_por_columna,
    'nulos_%': porcentaje_valores_nulos_por_columna,
    'nulos': valores_nulos_por_columna
})

# Mostrar el resultado
print(resultado_df)
```
